## <span style="color:#EF7C8E">Data cleaning:</span>

### <span style="color:#EF7C8E">1. Replace NaN with "No"</span>
* Integrated:
Is the graphics card integrated (replace NaN with "No", 14 rows)
* Dedicated:
Is the graphics card dedicated (replace NaN with "No", 14 rows)


```python
# REPLACE WITH NO:
# Replace all NaN with No because these columns have either a "Yes" or "No" Value
gpu.loc[:,['Dedicated','Integrated']]=gpu.loc[:,['Dedicated','Integrated']].fillna("No")
```

### <span style="color:#EF7C8E">2. Replace \n, \n-, and \t with NaN</span>

* all columns 

```python
# Replace all \n- and \n and \t with NaN
gpu=gpu.replace({'\n-':np.nan,'\n- ':np.nan,' \n-':np.nan,'\n':np.nan,' \n':np.nan,'\n ':np.nan,'\t':np.nan,' \t':np.nan,'\t ':np.nan})
```

### <span style="color:#EF7C8E">3. Replace NaN with Mode</span>

* Best_Resolution:
The best Screensize for reducing bottleneck to a minimum. (replace with mode: 642 rows)
* Direct_X:
What Direct X can the graphics card support (replace NaN with mode, 6 rows) 
* Memory:
The memory of graphics card (replace with mode: 420 rows)
* Memory_Bus:
The memory bus of graphics card (replace with mode or ffill: 62 rows)
* Memory_Type:
The memory type of the graphics card(replace with mode: 56 rows)
* Open_GL: 
What Open GL can the graphics card support.(replace with mode: 40 rows)
* Resolution_WxH:
Maximum supported resolution. (replace with mode: 195 rows)

```python
# REPLACE WITH MODE: 

# 1- Best_Resolution: 
gpu["Best_Resolution"].fillna(gpu["Best_Resolution"].mode()[0], inplace=True)
# 2- Direct_X
gpu["Direct_X"].fillna(gpu["Direct_X"].mode()[0], inplace=True)
# 3- Memory
gpu["Memory"].fillna(gpu["Memory"].mode()[0], inplace=True)
# 4- Memory_Bus
gpu["Memory_Bus"].fillna(gpu["Memory_Bus"].mode()[0], inplace=True)
# 5- Memory_Type
gpu["Memory_Type"].fillna(gpu["Memory_Type"].mode()[0], inplace=True)
# 6- Open_GL
gpu["Open_GL"].fillna(gpu["Open_GL"].mode()[0], inplace=True)
# 7- Resolution_WxH
gpu["Resolution_WxH"].fillna(gpu["Resolution_WxH"].mode()[0], inplace=True)
# 8- Max_Power
gpu["Max_Power"].fillna(method="ffill", inplace=True)
# 9- Core_Speed
gpu["Core_Speed"].fillna(method="ffill", inplace=True)
```

### <span style="color:#EF7C8E">4. Separate Column</span>
* Name: 
Name of the GPU. (divide first part as line, second specs/line-specifications)


```python

# Separate the Name into two new columns = Line and Line Description
gpu[["Line", "Line_Description"]] = gpu["Name"].str.split(" ", 1, expand=True)
# this resulted in one extra NaN so I dropped that row
gpu.dropna(subset=['Line_Description'], how='all', inplace=True)
```

### <span style="color:#EF7C8E">5. Date</span>
* Release_Date:
The GPU release date. (remove "\n" and "\n " and "\t" from the end, and "\nUnknown" to NaN and then replace with ffill and then convert to date format)

```python
# DATE CLEANUP 

# to remove the white spaces
gpu['Release_Date']=gpu['Release_Date'].map(lambda x: x.strip())

# Replace all "Unknown Release Date" with NaN
gpu=gpu.replace({'Unknown Release Date':np.nan}) # there were 28 unknowns
# Drop the NaNs there was 28 so it wouldnt affect that much
gpu.dropna(subset=['Release_Date'], how='all', inplace=True)
# Convert to Datetime format
gpu['Release_Date']=gpu['Release_Date'].map(lambda x: datetime.strptime(x, '%d-%b-%Y'))
```

### <span style="color:#EF7C8E">6. Drop Rows</span>
* Architecture: 
The GPU architecture used on the graphics card. (drop rows with null: 62 rows)

```python
#---------------------------------- drop rows ----------------------------------#

# drop the rows with empty value in column (Architecture) 
gpu.dropna(axis = 'index', how = 'all', subset = ['Architecture'], inplace = True)
```

### <span style="color:#EF7C8E">7. Replace Values with Null</span>

* Power Connector: is needed to run the graphics card(convert none to null)

```python
#---------------------------------- Replace values to null ----------------------------------#

# list with values needed to be replaced
none_val = ['None ', 'None  ']
# replace 'None ' and 'None  ' values with null  
gpu['Power_Connector'] = gpu['Power_Connector'].replace(none_val, np.NaN)
```

### <span style="color:#EF7C8E">8. Replace NaN Values With Zero</span>

* DVI_Connection: How many DVI connections does the graphics card have (replace NaN with 0, 750 rows)
* DisplayPort_Connection: How many DisplayPort connections does the graphics card have (replace NaN with 0, 2549 rows)
* HDMI_Connection: How many HDMI connections does the graphics card have (replace NaN with 0, 763 rows)
* ROPs: How many Raster Operation Pipelines does the GPU have (replace NaN with 0, 538 rows) 
* Shader: What Shader Model can the graphics card support.(replace NaN with 0, 107 rows)
* TMUs: How many texture mapping units does the GPU have.(replace NaN with 0, 538 rows)
* VGA_Connection: How many VGA connections does GPU have.(replace NaN with 0, 758 rows)

```python
#---------------------------------- Replace NaN values with zero ----------------------------------#

# list with all columns needed to replace Null value to zero 
to_zero = ['DVI_Connection', 'DisplayPort_Connection', 'HDMI_Connection','ROPs', 'Shader', 'TMUs', 'VGA_Connection']
# replace null value with zero 
gpu.update(gpu[to_zero].fillna(0))
```

```python
gpu['Power_Connector'].value_counts()
```

### <span style="color:#EF7C8E">9. Filling Null Values Using Forward Filling</span>

* Texture_Rate: How many texture pixels per second can the GPU produce.(replace NaN with ffill, 544 rows)
* Process: What nm technology does the GPU use. (replace with ffill: 463 rows)
* Pixel_Rate: The number of pixels per second the GPU can produce.(replace with ffill/mode: 544 rows)
* PSU: What Power Supply is needed to run the graphics card.(replace with ffill: 1175 rows)
* Memory_Speed: The memory speed of the graphic card's RAM(ffill method, 105 rows)
* Memory_Bandwidth: The memory bandwidth (ffill method, 121 rows)
* Power_Connector: Power Connector is needed to run the graphics card(convert none to null and ffill 2,363 rows)
* Boost_Clock: the MHz speed of the graphics card (ffill method, 1960 rows)
    

```python
#---------------------------------- filling null values using forward filling -----------------------------------#

# list with all columns needed to fill null using forward filling method
ffill_col = ['Texture_Rate', 'Process', 'Pixel_Rate', 'PSU', 'Memory_Speed', 'Memory_Bandwidth', 'Power_Connector', 'Boost_Clock']

# fill null value using forward filling method
gpu.update(gpu[ffill_col].fillna(method="ffill"))
```

### <span style="color:#EF7C8E">10. Filling Null Values Using Mode</span>

* Power_Connector: Power Connector is needed to run the graphics card(convert none to null and ffill 2,363 rows)
* Boost_Clock: the MHz speed of the graphics card (ffill method, 1960 rows)

```python
#---------------------------------- filling null values using mode -----------------------------------#
# finding mode value 
Power_Connector_mode = gpu['Power_Connector'].mode()
Boost_Clock_mode = gpu['Boost_Clock'].mode()

# fill null value using mode
gpu['Power_Connector'] = gpu['Power_Connector'].fillna(Power_Connector_mode)
gpu['Boost_Clock'] = gpu['Boost_Clock'].fillna(Boost_Clock_mode)
```

### <span style="color:#EF7C8E">11. Filling Null Values Using Backward Filling</span>

* Power_Connector: Power Connector is needed to run the graphics card(convert none to null and ffill 2,363 rows)
* Boost_Clock: the MHz speed of the graphics card (ffill method, 1960 rows)

```python
#---------------------------------- filling null values using backward filling -----------------------------------#
# list with all columns needed to fill null using backward filling method
bfill_col = ['Power_Connector', 'Boost_Clock']

# fill null value using backward filling method
gpu.update(gpu[bfill_col].fillna(method="bfill"))
```

### <span style="color:#EF7C8E">12. Filling Null in (Release_Price) column </span>

* Release_Price: The GPU price at time of release. (remove the '$', sub categorize it based on a metric/column and then fill the price with the mode of that metric/column *last step because its based on other columns) 

```python
#---------------------------------- price column -----------------------------------#

# removing '$' from value and any other characters if exists
gpu['Release_Price'] = gpu['Release_Price'].replace({'\$':''}, regex = True)

# convert column type to float 
gpu['Release_Price'] = gpu['Release_Price'].astype(float)

# fill prices based on memory  
gpu['Release_Price'] = gpu['Release_Price'].fillna(round(gpu.groupby('Memory')['Release_Price'].transform('mean'),2))

# fill prices based on memory type
gpu['Release_Price'] = gpu['Release_Price'].fillna(round(gpu.groupby('Memory_Type')['Release_Price'].transform('mean'),2))
```
