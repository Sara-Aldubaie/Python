```python
import pandas as pd
```

```python
# orignal link: https://archive.ics.uci.edu/ml/datasets/liver+disorders

PATH_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/liver-disorders/bupa.data"

# Save column names as a list
col_ls = ["mcv","alkphos", "sgpt", "sgot", "gammagt","drinks", "selector"]

# Pass in col_ls into names parameter to explicitly name each column
df_url = pd.read_csv(PATH_URL, names=col_ls)
df_url.head()


#Attribute information:
#   1. mcv: 	mean corpuscular volume
#   2. alkphos: 	alkaline phosphotase
#   3. sgpt: 	alamine aminotransferase
#   4. sgot:  	aspartate aminotransferase
#   5. gammagt: 	gamma-glutamyl transpeptidase
#   6. drinks: 	number of half-pint equivalents of alcoholic beverages drunk per day
#   7. selector:   field used to split data into two sets
```

```python
# original link: https://archive.ics.uci.edu/ml/machine-learning-databases/letter-recognition/

PATH_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/letter-recognition/letter-recognition.data"

# Save column names as a list
col_ls = ["lettr","x-box","y-box","width","high","onpix","x-bar","y-bar","x2bar","y2bar","xybar","x2ybr","xy2br","x-ege", "xegvy", "y-ege", "yegvx"]

# Pass in col_ls into names parameter to explicitly name each column
df_url = pd.read_csv(PATH_URL, names=col_ls)
df_url.head()


#Attribute information:
# 1. lettr capital letter (26 values from A to Z)
# 2. x-box horizontal position of box (integer)
# 3. y-box vertical position of box (integer)
# 4. width width of box (integer)
# 5. high height of box (integer)
# 6. onpix total number on pixels (integer)
# 7. x-bar mean x of on pixels in box (integer)
# 8. y-bar mean y of on pixels in box (integer)
# 9. x2bar mean x variance (integer)
# 10. y2bar mean y variance (integer)
# 11. xybar mean x y correlation (integer)
# 12. x2ybr mean of x * x * y (integer)
# 13. xy2br mean of x * y * y (integer)
# 14. x-ege mean edge count left to right (integer)
# 15. xegvy correlation of x-ege with y (integer)
# 16. y-ege mean edge count bottom to top (integer)
# 17. yegvx correlation of y-ege with x (integer)
```
